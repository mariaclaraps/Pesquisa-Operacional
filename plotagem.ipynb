```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rotas = pd.read_csv("rotas.csv")

# Coordenadas (iguais s do .dat)
coords = {
    0: (50, 0),
    1: (12, 38),
    2: (44, 81),
    3: (55, 22),
    4: (23, 65),
    5: (70, 50),
    6: (15, 10),
    7: (90, 85),
    8: (60, 70),
    9: (5, 55),
    10: (40, 20),
    11: (80, 10),
    12: (30, 40),
    13: (50, 50),
    14: (10, 80),
    15: (85, 45),
}



plt.figure(figsize=(10, 8))

for k in rotas["veiculo"].unique():

    dados_v = rotas[rotas["veiculo"] == k]

    # distncia total do veculo (soma das arestas)
    dist_total = dados_v["distancia"].sum()

    for _, row in dados_v.iterrows():
        i, j = row["origem"], row["destino"]

        xi, yi = coords[i]
        xj, yj = coords[j]

        plt.plot([xi, xj], [yi, yj], marker="o", label=f"Veculo {k}" if i == 0 else "")


# Plotar clientes
for v, (x, y) in coords.items():
    if v == 0:
        plt.scatter(x, y, s=150)
        plt.text(x + 1, y + 1, "Depsito", fontsize=10)
    else:
        plt.scatter(x, y)
        plt.text(x + 1, y + 1, str(v), fontsize=9)

plt.title("Rotas dos Veculos")
plt.xlabel("Coordenada X")
plt.ylabel("Coordenada Y")
plt.grid(True)
plt.legend()
plt.show()
```

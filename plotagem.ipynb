```python
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# DADOS
# ==============================
rotas = pd.read_csv("rotas.csv")

coords = {
    0:(50,0), 1:(12,38), 2:(44,81), 3:(55,22), 4:(23,65),
    5:(70,50), 6:(15,10), 7:(90,85), 8:(60,70), 9:(5,55),
    10:(40,20), 11:(80,10), 12:(30,40), 13:(50,50),
    14:(10,80), 15:(85,45)
}

# ==============================
# PLOT
# ==============================
plt.figure(figsize=(10,8))

# Plotar clientes
for i, (x,y) in coords.items():
    if i == 0:
        plt.scatter(x, y, s=200, marker="s")
        plt.text(x+1, y+1, "Depsito", fontsize=10, weight="bold")
    else:
        plt.scatter(x, y)
        plt.text(x+1, y+1, str(i), fontsize=9)

# Plotar rotas por veculo
for k in rotas["veiculo"].unique():
    r = rotas[rotas["veiculo"] == k]
    for _, row in r.iterrows():
        x1, y1 = coords[row["origem"]]
        x2, y2 = coords[row["destino"]]
        plt.plot([x1,x2], [y1,y2], linewidth=2)

plt.title("Rotas do CVRP")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```
